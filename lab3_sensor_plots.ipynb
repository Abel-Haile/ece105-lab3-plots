```python
# Generate simulated temperature readings from two sensors.
# Sensor A: mean 25 C, std 3 C, 200 readings.
# Sensor B: mean 27 C, std 4.5 C, 200 readings.
# Also generate 200 timestamps uniformly from 0 to 10 seconds.
# Use np.random.default_rng with a seed = last 4 digits of your Drexel ID.

import numpy as np
import matplotlib.pyplot as plt

seed = 2493
rng = np.random.default_rng(seed)
n = 200

# Timestamps uniformly in [0, 10], sorted so plots show a time series
t = rng.uniform(0, 10, n)
t.sort()

# Sensor readings per spec
sensor_a = rng.normal(loc=25.0, scale=3.0, size=n)
sensor_b = rng.normal(loc=27.0, scale=4.5, size=n)
```

```python
sensor_a.shape
```

```python
# Scatter plot: sensor reading (y-axis) vs timestamp in seconds (x-axis).
# Color Sensor A points blue and Sensor B points orange.
# Add axis labels with units, a title, and a legend.
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 5))
plt.scatter(t, sensor_a, color='blue', alpha=0.7, label='Sensor A')
plt.scatter(t, sensor_b, color='orange', alpha=0.7, label='Sensor B')
plt.xlabel('Time (s)')
plt.ylabel('Temperature (°C)')
plt.title('Simulated Temperature Readings from Two Sensors')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Overlaid histogram of Sensor A and Sensor B temperature distributions.
# Use 30 bins, alpha=0.5 for transparency so both distributions are visible.
# Add vertical dashed lines at each sensor's mean.
# Include a legend labeling each sensor.
```

```python
# Overlaid histogram of Sensor A and Sensor B temperature distributions.
# Use 20 bins, semi-transparent bars, and show mean lines.
import matplotlib.pyplot as plt
plt.figure(figsize=(8,4))
bins = 20
plt.hist(sensor_a, bins=bins, alpha=0.6, label=f'Sensor A (μ={sensor_a.mean():.2f}, σ={sensor_a.std():.2f})', color='C0')
plt.hist(sensor_b, bins=bins, alpha=0.5, label=f'Sensor B (μ={sensor_b.mean():.2f}, σ={sensor_b.std():.2f})', color='C1')
plt.axvline(sensor_a.mean(), color='C0', linestyle='--')
plt.axvline(sensor_b.mean(), color='C1', linestyle='--')
plt.xlabel('Temperature (°C)')
plt.ylabel('Count')
plt.title('Histogram of Sensor Temperatures')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Side-by-side box plot comparing Sensor A and Sensor B distributions.
# Label x-axis with sensor names, y-axis with "Temperature (deg C)".
# Add a horizontal dashed line at the overall mean of both sensors combined.
```

```python
# Box plot comparing Sensor A and Sensor B.
# Show medians, means, notches and outliers.
import matplotlib.pyplot as plt
plt.figure(figsize=(6,4))
data = [sensor_a, sensor_b]
labels = ['Sensor A', 'Sensor B']
bp = plt.boxplot(data, labels=labels, notch=True, showmeans=True, patch_artist=True)
# Color the boxes separately
colors = ['C0', 'C1']
for patch, color in zip(bp['boxes'], colors):
    patch.set_facecolor(color)
    patch.set_alpha(0.6)
# Style medians and means
for median in bp.get('medians', []):
    median.set(color='black', linewidth=1.5)
if 'means' in bp:
    for mean in bp['means']:
        mean.set(marker='D', markeredgecolor='black', markerfacecolor='gold')
plt.ylabel('Temperature (°C)')
plt.title('Boxplot of Sensor Temperatures')
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()

plt.figure(figsize=(8, 5))
plt.boxplot([sensor_a, sensor_b], labels=['Sensor A', 'Sensor B'])

overall_mean = np.mean(np.concatenate([sensor_a, sensor_b]))
plt.axhline(overall_mean, color='red', linestyle='--', linewidth=2, label=f'Overall mean ({overall_mean:.2f} °C)')

plt.xlabel('Sensor')
plt.ylabel('Temperature (deg C)')
plt.title('Distribution Comparison of Sensor A and Sensor B')
plt.legend()
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```
